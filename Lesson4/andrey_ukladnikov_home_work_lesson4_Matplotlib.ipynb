# Задание 2

С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10
включительно.
Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для
координат по вертикали. Линия графика должна быть зеленого цвета.
Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси -
'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.

```python
t = np.linspace(0, 10, 51,endpoint=True)
```

```python
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
legend_font = {
    "size": 7,
    "family": "serif",
}
label_font = {
    "fontsize": 9,
    "family": "serif",
}
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "black",
    "family": "serif",
}

plt.plot(t, f, label="cos(x)", color="green", linestyle="solid")
plt.title("График f(t)", fontdict=title_font)
plt.xlabel("Значения t", fontdict=label_font)
plt.ylabel("Значения f", fontdict=label_font)
plt.axis([0.5, 9.5, -2.5, 2.5])
legend = plt.legend(loc="lower right", prop=legend_font)
plt.setp(legend.get_texts(), color="DarkBlue")
plt.show()
```

# *Задание 3

С помощью функции linspace библиотеки Numpy создайте массив x из 51 числа от -3 до 3
включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с
названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных
графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется
для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте
y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.Дайте название графикам:
'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3, 3, 51,endpoint=True)
```

```python
print(x)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title("График y1")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title("График y2")

ax3.plot(x, y3)
ax3.set_title("График y3")

ax4.plot(x, y4)
ax4.set_title("График y4")
```

```python
data_base=pd.read_csv('creditcard.csv')
```

```python
data=data_base.copy()
data.shape
```

```python
plt.style.use('fivethirtyeight')
```

```python
normal_tranc_count = data.loc[data.Class == 0,'Class'].value_counts()
```

```python
print(normal_tranc_count[0])
```

```python
fraudulent_tranc_count = data.loc[data.Class == 1,'Class'].value_counts()
```

```python
print(fraudulent_tranc_count[1])
```

```python
transaction_counts = normal_tranc_count.append(fraudulent_tranc_count)
```

```python
transaction_types = ["normal", "fraudulent"]
```

```python
transaction_counts.index = transaction_types
```

```python
print(transaction_counts)
```

```python
transaction_counts.plot(kind="barh")
plt.show()
```

```python
transaction_counts.plot(kind="barh",logx=True)
plt.show()
```

```python
fraudulent_v1 = data.loc[data.Class == 1,'V1']
```

```python
normal_v1 = data.loc[data.Class == 0,'V1']
```

```python
legend_font = {
    "size": 7,
    "family": "serif",
}
label_font = {
    "fontsize": 9,
    "family": "serif",
}
plt.hist(fraudulent_v1,bins=20, orientation="vertical",alpha=0.5,label="Class 1",density=True, color="red", ec="white")
plt.hist(normal_v1,bins=20, orientation="vertical",alpha=0.5, label="Class 0",density=True,color="lightgrey", ec="white")
legend = plt.legend(loc="upper left", prop=legend_font)
plt.setp(legend.get_texts(), color="DarkBlue")
plt.xlabel("Class", fontdict=label_font)
plt.show()
```
